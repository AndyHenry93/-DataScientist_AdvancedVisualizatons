# College Degree Gender Gap

The purpuse of this project is to show the rate of change within college degrees of Males and Females. In this project I wanted to portry my visualzation skills by making the graphs easily readable and understandable, I also wanted to make them accessible to all so the only colors used in this project are from the Tableau RGB color blind  Palette.    

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
women_degrees = pd.read_csv('./Dataset/percent-bachelors-degrees-women-usa.csv')

cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 
             'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 
                 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 
              'Agriculture','Business', 'Architecture']
```

```python
fig = plt.figure(figsize=(16, 20))

## Generate first column of line charts. STEM degrees.
for i in range(0,18,3):
    Stem_idex = int(i/3)
    ax = fig.add_subplot(6,3,i+1)
#     Stem Degree category
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[Stem_idex]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[Stem_idex]], c=cb_orange, label='Men', linewidth=3)
    for key,value in ax.spines.items():
        value.set_visible(False)
    ax.tick_params(bottom = False, top = False, left= False, right = False,labelbottom = False)
    ax.set_yticks([0,100])
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[Stem_idex])
    ax.axhline(y = 50, c = (171/255,171/255,171/255), alpha = 0.3)
    if Stem_idex == 0:
        ax.text(2005, 87, 'Women')
        ax.text(2002, 8, 'Men')
    elif Stem_idex == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        
ax.tick_params(labelbottom = True)
        
## Generate Second column of line charts. Liberal Art degrees.
for i in range(1,16,3):
    Lib_index = int((i-1)/3)
    ax = fig.add_subplot(6,3,i+1)
#     Liberal Art Degree category
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[Lib_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[Lib_index]], c=cb_orange, label='Men', linewidth=3)
    for key, value in ax.spines.items():
        value.set_visible(False)
    ax.tick_params(bottom = False, top = False, left= False, right = False,labelbottom = False)
    ax.set_yticks([0,100])
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[Lib_index])
    ax.axhline(y = 50, c = (171/255,171/255,171/255), alpha = 0.3)
    if Lib_index == 0:
        ax.text(2005, 87, 'Women')
        ax.text(2002, 8, 'Men')
        
ax.tick_params(labelbottom = True)
        
## Generate Third column of line charts. Other degrees.
for i in range(2,20,3):
    other_index = int((i-2)/3)
    ax = fig.add_subplot(6,3,i+1)
#     Other Degree category
    ax.plot(women_degrees['Year'], women_degrees[other_cats[other_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[other_index]], c=cb_orange, label='Men', linewidth=3)
    for key, value in ax.spines.items():
        value.set_visible(False)
    ax.tick_params(bottom = False, top = False, left= False, right = False,labelbottom = False)
    ax.set_yticks([0,100])
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[other_index])
    ax.axhline(y = 50, c = (171/255,171/255,171/255), alpha = 0.3)
    if other_index == 0:
        ax.text(2005, 87, 'Women')
        ax.text(2002, 8, 'Men')
    elif other_index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        
ax.tick_params(labelbottom = True)

# creating a layout so the graph is scaled better on laptops 
plt.tight_layout()

# saving the figure
plt.savefig("Gender_degrees.png",bbox_inches='tight',dpi=300)
        
plt.show()
```

First when looking at degrees in the STEM category, we see that some gaps are gradualy decresing over time and others still being more dominate by either males or females. Other than foreign languages, we see all other major rather gender neutral and diverse.<br>
